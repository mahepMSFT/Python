```python
class Student:
    def __init__(self, name, rollno, course):
        self.name = name
        self.rollno = rollno
        self.course = course

    def display(self):
        print("Student Name:", self.name)
        print("Roll No:", self.rollno)
        print("Course:", self.course)

s1 = Student("Mahender", 101, "Python Full Stack")
s1.display()
```

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

a = Animal()
d = Dog()

a.sound()
d.sound()
```

```python
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Please enter valid integers.")
```

```python
class Student:
    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks

    def display_details(self):
        print("Name:", self.name)
        print("Roll No:", self.rollno)
        print("Marks:", self.marks)

student1 = Student("Sreya", 102, 95)
student1.display_details()
```

```python
class Father:
    def skills(self):
        print("Father: Gardening")

class Mother:
    def talents(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    def abilities(self):
        print("Child: Learning both skills")

c = Child()
c.skills()
c.talents()
c.abilities()
```
